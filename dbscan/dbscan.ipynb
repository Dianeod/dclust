## DBSCAN Clustering

Density-based spatial clustering of applications with noise (DBSCAN) is a density based clustering algorithm which groups together points that are closely packed together in areas of high density. Any points in low density regions are set as outliers. C:

1.minpts: minimum points in a given neighbourhood for it to be classified as a cluster

2.epsilon: Measures the distance of the neighbourhood using radius epsilon

The benefit of this method of clustering is that it deals well with outliers and clusters fast. However it doesn't deal very well with clusters of varying densities.

The dbscan formulas is contained within db.q

```q
\l db.q
ds:.p.import[`sklearn.datasets]
```

### Sample 1

```q
moon:ds[`:make_moons][`n_samples pykw 2000;`noise pykw 0.05]`
```

```q
plt:.p.import`matplotlib.pyplot
plt[`:scatter][;]. flip sample:moon[0];
plt[`:show][];
```

```q
\ts kmoon:dbscan[moon[0];6;0.05]
```

```q
plt:.p.import[`matplotlib]`:pyplot;
{plt[`:scatter][;]. flip x y}[moon[0]]each kmoon;
plt[`:show][];
```

### Sample 2

```q
dsc:{[n;r;s](r+n?s-r)*/:(cos;sin)@\:4*n?acos 0}
data :flip(-5 1)+(1 1.5)*dsc[1000; 0;1.8],'dsc[2000;3.1;4.2],'dsc[4000;5.2;6.5]
data,:flip(4 -1)+(1 8)*dsc[4000; 0;1.]
sample:data@:neg[4000&count data]?count data
```

```q
plt:.p.import`matplotlib.pyplot
plt[`:scatter][;]. flip sample;
plt[`:show][];
```

```q
\ts kk:dbscan[sample;2;1]
```

```q
plt:.p.import[`matplotlib]`:pyplot;
{plt[`:scatter][;]. flip x y}[sample]each kk;
plt[`:show][];
```
